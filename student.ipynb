```python
#Plotting a line plot to display the trend of average ratings over the years (using  "start_year" and "averagerating" columns
# Merge the two DataFrames based on the common index 'tconst'
merged_df = pd.merge(title_ratings, title_basics[['start_year']], left_index=True, right_index=True)

# Group the data by 'start_year' and calculate the mean of 'averagerating' for each year
average_ratings_by_year = merged_df.groupby('start_year')['averagerating'].mean()

# Plot the line plot
plt.figure(figsize=(10, 6))
plt.plot(average_ratings_by_year.index, average_ratings_by_year.values, marker='o', color='b')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Trend of Average Ratings Over the Years')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
